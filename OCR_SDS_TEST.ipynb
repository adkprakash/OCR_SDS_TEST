```python
import numpy as np
import pandas as pd
import cv2

import pytesseract

import pathlib
import os
import matplotlib.pyplot as plt
import seaborn as sns
```

Loading images from folders

```python
imgdata=[]
def loadimg(path):
 for imgname in os.listdir(path):
    img = cv2.imread(os.path.join(path, imgname))
    if img is not None:
        imgdata.append(img)
```

```python
path=['/content/drive/MyDrive/MachineLearningEngineerTest/ML/data/images/0','/content/drive/MyDrive/MachineLearningEngineerTest/ML/data/images/2',
      '/content/drive/MyDrive/MachineLearningEngineerTest/ML/data/images/4','/content/drive/MyDrive/MachineLearningEngineerTest/ML/data/images/6',
      '/content/drive/MyDrive/MachineLearningEngineerTest/ML/data/images/9']
for str in path:
  loadimg(str)
```
